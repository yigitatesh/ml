### K-Nearest Neighbour Classifier

```python
class KNNClassifier:
    """K-Nearest Neighbour Classifier."""
    
    def __init__(self, n_neighbors=1):
        self.n_neighbors = n_neighbors
    
    def fit(self, X, y):
        """Fits classifier with train sets."""
        self.X_train = X
        self.y_train = y
        return self
    
    def predict(self, X_test):
        """Make predictions for test dataset.
        
        Parameters
        ----------
        X_test : dataset for predictions
            (n_samples, n_features)
        
        Returns
        -------
        array : a numpy array contains prediction(s)"""
        # predict outputs
        predictions = []
        for row in X_test:
            predictions.append(self._predict(row)[0])
            
        return np.array(predictions)
    
    def _predict(self, test_row):
        """Make prediction for a test row.
        
        Parameters
        ----------
        X_test : row for prediction
            (n_features)
        
        Returns
        -------
        array : a numpy array contains the prediction
            example: np.array([1])"""
        # find distances and y values
        # sum(axis=1) is columnwise sum 
        distances = ((self.X_train - test_row) ** 2).sum(axis=1) ** 0.5
        # sort by distances
        sorted_indices = distances.argsort()
        # predictions
        predictions = self.y_train[sorted_indices][:self.n_neighbors]
        
        return mode(predictions)[0]
    
    def score(self, X_test, y_test):
        """Calculate accuracy by scoring.
        accuracy = True predictions / All predictions"""
        predictions = self.predict(X_test)
        return np.mean(predictions == y_test)
```

### K-Nearest Neighbor Regressor

```python
class KNNRegressor:
    """K-Nearest Neighbor Regressor"""
    def __init__(self, n_neighbors=1):
        self.n_neighbors = n_neighbors
    
    def fit(self, X, y):
        """Fits classifier with train sets.
        
        Parameters
        ----------
        X: features dataset for train
            (n_samples, n_features)
        y: output values for train
            (n_samples)"""
        self.X_train = X
        self.y_train = y
        return self
    
    def predict(self, X_test):
        """Make predictions for test dataset.
        
        Parameters
        ----------
        X_test : dataset for predictions
            (n_samples, n_features)
        
        Returns
        -------
        array : a numpy array contains prediction(s)"""
        # find distances and y values
        predictions = []
        for row in X_test:
            predictions.append(self._predict(row))
            
        return np.array(predictions)
    
    def _predict(self, test_row):
        """Make prediction for a test row.
        
        Parameters
        ----------
        X_test : row for prediction
            (n_features)
        
        Returns
        -------
        array : a numpy array contains the prediction"""
        # find distances and y values
        distances = ((self.X_train - test_row) ** 2).sum(axis=1) ** 0.5
        # sort by distances
        sorted_indices = distances.argsort()
        # predictions
        predictions = self.y_train[sorted_indices][:self.n_neighbors]
        
        return np.mean(predictions)
    
    def score(self, X_test, y_test):
        """Calculate R^2 score
        R^2 = 1 - u/v
        u is the regression sum of squares
            ((y_test - y_pred) ** 2).sum()
        v is the residual sum of squares
            ((y_test - y_test.mean()) ** 2).sum()
            
        Parameters
        ----------
        X_test: dataset for predictions
            (n_samples, n_features)
        y_test: true output values for X_test
        
        Returns
        -------
        score: R^2 score"""
        y_pred = self.predict(X_test)
        u = ((y_test - y_pred) ** 2).sum()
        v = ((y_test - y_test.mean()) ** 2).sum()
        
        return 1 - u/v
```

### TEST

```python
test_scores = []
train_scores = []
k_list = [1, 3, 5, 7, 9, 11, 13, 15]
random_states = np.random.randint(0, 100, size=20)
n_tests = len(random_states)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0, test_size=0.25)
for k in k_list:
    knn = KNNClassifier(k)
    train_sum = 0
    test_sum = 0
    for r in random_states:
        X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=r, test_size=0.25)
        knn.fit(X_train, y_train)
        train_sum += knn.score(X_train, y_train)
        test_sum += knn.score(X_test, y_test)
    test_scores.append(test_sum / n_tests)
    train_scores.append(train_sum / n_tests)
```

```python
fig, ax = plt.subplots()

ax.plot(k_list, test_scores, color="blue", label="test scores")
ax.plot(k_list, train_scores, color="green", label="train scores")

plt.legend()
```
